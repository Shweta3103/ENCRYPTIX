```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card = pd.read_csv('E:\DATA Science\Internship\creditcard.csv')
```

```python
# checking for missing values
credit_card.isnull().sum()
```

```python
# distribution of legit transaction &  fraudent transactions

credit_card['Class'].value_counts()
```

```python
## 0 - Normal Transaction, 1- fraudent transaction
# separating the data for analysis

legit = credit_card[credit_card.Class == 0]
fraud = credit_card[credit_card.Class == 1]
print(legit.shape)
print(fraud.shape)
```

```python
# statistic data measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#Compare the values for both transactons

credit_card.groupby('Class').mean()
```

```python
# Under Sampling
# Build a Sample dataset containing similar distribution of normal transactions and fraudlent transactions
# Number of Fraudalent Transaction - 492

legit_sample = legit.sample(n=492)
```

```python
# Concatenating two DataFrames

new_dataset = pd.concat([legit_sample, fraud], axis=0)
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#splitting the data into features $ targets

X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Split the data into Training data & Testing Data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Traininh
Logistic Regression

```python
model = LogisticRegression()
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=3000)
model.fit(X_train,Y_train)
```

```python
#training the logistic Regression Model with Training Data

model.fit(X_train, Y_train)
```

```python
# Model Evaluation

#Accuracy Score

#accuracy on training data

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training Data: ', training_data_accuracy)
```

```python
# accuracy on test data

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data: ', test_data_accuracy)
```
